```python
import numpy as np
import pandas as pd
import keras

#importing Keras, Library for deep learning 
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D

from sklearn.model_selection import train_test_split
```

```python
#LOAD DATA
from sklearn.datasets import load_digits
digits = load_digits()
#DATAFRAMING
df_y = pd.DataFrame(data=digits.target).iloc[:,0]
df_x = pd.DataFrame(data=digits.data, columns=digits.feature_names)
```

```python
df_x.head(5)
```

```python
df_y = np.array(df_y)
print(df_y)
```

```python
#reshaping data to 8x8 array
#df_x.iloc[0].values.reshape(8,8).astype('uint8')
```

```python
new_df_x = digits.data.reshape(len(digits.data),8,8,1).astype('uint8')

new_df_x = np.array(new_df_x)

print(new_df_x)
```

```python
#PREPROCESSING DATA
#df_x = np.array(df_x)
#print(df_x)
#df_x.shape
```

```python
new_df_y = keras.utils.to_categorical(df_y,num_classes=10)
new_df_y = np.array(new_df_y)
```

```python
print(df_y[57])
print(new_df_y[57])
```

```python
x_train, x_test, y_train, y_test = train_test_split(new_df_x, new_df_y, test_size=0.2, random_state=4)
```

```python
#CNN MODEL
model = Sequential()
model.add(Convolution2D(16,3,data_format='channels_last',activation='relu',input_shape=(8,8,1)))
model.add(MaxPooling2D(pool_size=(1,1)))
model.add(Flatten())
model.add(Dense(100))
model.add(Dropout(0.5))
model.add(Dense(10))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adadelta',metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test))
```
